```python
#Import the depencies
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("C:/Users/ADMIN/Downloads/pandas.png")
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.title("Panda")
plt.imshow(img_rgb)
plt.show()
```

```python
gray_image =cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.title("GRAY IMAGE")
plt.imshow(gray_image, cmap='gray')
plt.show()
```

```python
invert_image = cv2.bitwise_not(gray_image)
plt.title("Inverted Image")
plt.imshow(invert_image, cmap='gray')
plt.show()
```

```python
img_Blur = cv2.GaussianBlur(invert_image,(111,111),0)
plt.title("Blurred Image")
plt.imshow(img_Blur, cmap='gray')
plt.show()
```

```python
inv_blur= cv2.bitwise_not(img_Blur)
plt.title("Inverted Blur Image")
plt.imshow(inv_blur, cmap='gray')
plt.show()
```

```python
sketch=cv2.divide(gray_image, inv_blur, scale= 256.0)
plt.title("Sketch Image")
plt.imshow(sketch, cmap='gray')
plt.show()
```
